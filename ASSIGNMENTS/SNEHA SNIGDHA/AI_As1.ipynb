```python
#Task 1
import numpy as np
import pandas as pd

np.random.seed(42)

# Create a NumPy array with random values
data = np.random.randn(4, 4)

# Create the DataFrame
df = pd.DataFrame(data)

df
```

```python
#Task 2
df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
df
```

```python
#Task 3
df_stats = df.describe()
df_stats
```

```python
#Task 4
null_values = df.isnull().sum()
null_values
```

```python
#Task 5
# Using index location method
column_2_idx = df['Random value 2']
column_3_idx = df['Random value 3']

# Printing the results
print("\nColumn 'Random value 2' with index location method:\n", column_2_idx)
print("\nColumn 'Random value 3' with index location method:\n", column_3_idx)
```
